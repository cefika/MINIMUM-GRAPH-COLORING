```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def parse_file_info(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    
    filenames1 = []
    broj_boja_1 = []
    vreme_1 = []
    
    for i in range(0, len(lines), 3):
        filename_line = lines[i].strip()
        broj_boja_line = lines[i + 1].strip()
        vreme_line = lines[i + 2].strip()
        
        # Uzimamo ime fajla
        if filename_line.endswith('.txt'):
            filenames1.append(filename_line)
        
        # Uzimamo broj boja
        parts = broj_boja_line.split(': ')
        broj_boja_1.append(int(parts[1]))
        
        # Uzimamo vreme
        parts = vreme_line.split(': ')
        vreme_1.append((parts[1].split(' ')[0]))
    
    return filenames1, broj_boja_1, vreme_1

# Primer korišćenja
filename = 'Rezultati/aco.txt'
filenames1, broj_boja_1, vreme_1 = parse_file_info(filename)
vreme_1 = [float(v.replace('sekundi', '')) for v in vreme_1]

filename = 'Rezultati/pso.txt'
filenames2, broj_boja_2, vreme_2 = parse_file_info(filename)
vreme_2 = [float(v.replace('sekundi', '')) for v in vreme_2]

filename = 'Rezultati/brute_force.txt'
filenames2, broj_boja_3, vreme_3 = parse_file_info(filename)
vreme_3 = [float(v.replace('sekundi', '')) for v in vreme_3]

filename = 'Rezultati/genetic_algorithm.txt'
filenames2, broj_boja_4, vreme_4 = parse_file_info(filename)
vreme_4 = [float(v.replace('sekundi', '')) for v in vreme_4]

filename = 'Rezultati/simulated_annealing.txt'
filenames2, broj_boja_5, vreme_5 = parse_file_info(filename)
vreme_5 = [float(v.replace('sekundi', '')) for v in vreme_5]
```

```python
plt.figure(figsize=(4, 4))
plt.scatter(broj_boja_1,vreme_1, color='blue')
plt.xlabel('Broj boja')
plt.ylabel('Vreme (sekunde)')
plt.title('Broj boja vs Vreme aco')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(4, 4))
plt.scatter(broj_boja_4,vreme_4, color='blue')
plt.xlabel('Broj boja')
plt.ylabel('Vreme (sekunde)')
plt.title('Broj boja vs Vreme genetic_algorithm')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(4, 4))
plt.scatter(broj_boja_5,vreme_5, color='blue')
plt.xlabel('Broj boja')
plt.ylabel('Vreme (sekunde)')
plt.title('Broj boja vs Vreme simulated_annealing')
plt.grid(True)
plt.show()
```

```python
# Funkcija za crtanje grafikona za svaki fajl
def plot_file_data(ime_fajla, broj_boja_1, broj_boja_2, broj_boja_3, broj_boja_4, broj_boja_5, vreme_1, vreme_2, vreme_3, vreme_4, vreme_5):
    indeks = filenames2.index(ime_fajla)
    boja_1 = broj_boja_1[indeks]
    boja_2 = broj_boja_2[indeks]
    boja_3 = broj_boja_3[indeks]
    boja_4 = broj_boja_4[indeks]
    boja_5 = broj_boja_5[indeks]

    vreme_1_fajl = vreme_1[indeks]
    vreme_2_fajl = vreme_2[indeks]
    vreme_3_fajl = vreme_3[indeks]
    vreme_4_fajl = vreme_4[indeks]
    vreme_5_fajl = vreme_5[indeks]

    # Pravljenje grafikona
    fig, ax1 = plt.subplots()

    # Prvi podatak (vreme izvršavanja)
    ax1.set_xlabel('Fajlovi')
    ax1.set_ylabel('Vreme izvršavanja (sekunde)', color='tab:blue')
    ax1.bar([0, 1, 2, 3, 4], [vreme_3_fajl, vreme_1_fajl, vreme_2_fajl, vreme_4_fajl, vreme_5_fajl], color='tab:blue')
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    # Drugi podatak (broj boja)
    ax2 = ax1.twinx()
    ax2.set_ylabel('Broj boja', color='tab:red')
    ax2.plot([0, 1, 2, 3, 4], [boja_3, boja_1, boja_2, boja_4, boja_5], color='tab:red', marker='o')
    ax2.tick_params(axis='y', labelcolor='tab:red')

    # Naslovi i podešavanja
    plt.title(f'Podaci za {ime_fajla}')
    plt.xticks([0, 1, 2, 3, 4], ['Brute_Force','ACO', 'PSO', "GA", 'Simulirano kaljenje'])
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    fig.tight_layout()

    plt.show()

# Prikazivanje grafikona za svaki fajl
for ime in filenames2:
    plot_file_data(ime, broj_boja_1, broj_boja_2, broj_boja_3, broj_boja_4, broj_boja_5, vreme_1, vreme_2, vreme_3, vreme_4, vreme_5)
```
